```python
import numpy as np
import pandas as pd
```

```python
id_df = pd.read_csv('DATA/id.csv')
id_df.head()
```

```python
games_df = pd.read_csv('DATA/name.csv')
games_df.head()
```

```python
ratings_df = pd.read_csv('DATA/rating.csv')
ratings_df.head()
```

```python
pub_df = pd.read_csv('DATA/publisher.csv')
pub_df.head()
```

```python
df = games_df.merge(ratings_df, on='GameID')
df
```

```python
G_j = 'Mario Kart DS'
recommended_games = []


game_db = df[df['Name_x'] == G_j]\
            .sort_values(by='rating', ascending=False)


for user in game_db.iloc[:5]['userID'].values:
    
    rated_games = df[df['userID'] == user]
    
    rated_games = rated_games[rated_games['Name_x'] != G_j]\
                    .sort_values(by='rating', ascending=False)\
                    .iloc[:5]
    

    recommended_games.extend(list(rated_games['Name_x'].values))
    
recommended_games = np.unique(recommended_games)
    
for game in recommended_games:
    print(game)
```

```python
ggame_genres = df[df['Name_x'] == G_j].iloc[0]['Genre_x'].split('|')
scores = {}

for game in recommended_games:
    gamed = df[df['Name_x'] == game].iloc[0]
    game_genres = gamed['Genre_x'].split('|')
    score = 0
    

    for ggame_genre in ggame_genres:
        if ggame_genre in game_genres:
            score += 1
    
    scores[game] = score
    
recommended_games = sorted(scores, key=lambda x: scores[x])[::-1]  
```

```python
for game in recommended_games:
    print(game)
```
